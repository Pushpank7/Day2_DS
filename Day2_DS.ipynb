# **Pivot Table in Pandas - Sales Analysis by Region and Product**

```python
import pandas as pd

data = {
    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450],
    'Quantity': [5, 10, 15, 20, 25, 30, 35, 40]
}

df = pd.DataFrame(data)
df
```

# **Pivot Table - Total Sales by Region and Product**

```python
pivot_sales = pd.pivot_table(df, values = 'Sales', index = 'Region', columns ='Product',aggfunc = 'sum',fill_value=0)
pivot_sales
```

# **Pivot Table - Average Quantity by Region and Product**

```python
pivot_qty = pd.pivot_table(df, values = 'Quantity', index = 'Region', columns ='Product',aggfunc = 'mean',fill_value=0)
pivot_qty
```

# **Crosstab - Product Sales by Region**

```python
data = {
    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
}
df = pd.DataFrame(data)
df
```

# **Coount Product Appearances in Each Region**

```python
cross = pd.crosstab(df['Region'], df['Product'])
cross
```
